```python
import pandas as pd
import numpy as np
import json
```

```python
def parseReqFile(file):
    with open(file,'r') as rFile:
        lines = rFile.readlines()
    k = [json.loads(line) for line in lines]
    return k

j = parseReqFile("crawl/hindilinks4u.kim/requests.json")
```

```python
j[8]
```

```python
for idj, jj in enumerate(j):
    print(jj['http_req'])
```

```python
for idj, jj in enumerate(j):
    stack = jj['call_stack']
    if 'stack' in stack.keys():
        stack = stack['stack']
        if 'callFrames' in stack:
            for cf in stack['callFrames']:
                print("====={}====={}====={}".format(idj, cf['functionName'], cf['url']))
    
```
